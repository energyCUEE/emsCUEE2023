# Categorize Load and Solar

## Load

### Pload < 4 kW => low_load
### 4 kW <= Pload < 6 kW => medium_load
### 6 kW <= Pload => high_load

## Solar

### Ppv <= 1.5 kW => low_solar
### 1.5 kW < Ppv => high_solar

```python
def categorize_load(Pload):
    left_pivot = 4
    right_pivot = 6
    if Pload < left_pivot:
        load_type = 'low_load'
    elif left_pivot <= Pload and Pload < right_pivot:
        load_type = 'medium_load'
    elif right_pivot <= Pload:
        load_type = 'high_load'
        
    return load_type
```

```python
def categorize_pv(Ppv):
    pivot = 1.5
    if Ppv <= pivot:
        solar_type = 'low_solar'
    elif pivot < Ppv:
        solar_type= 'high_solar'
        
    return solar_type
```

```python
df_load_mean['load_type'] = df_load_mean['Ptot (kW)'].apply(categorize_load)
df_pv_mean['solar_type'] = df_pv_mean['Ptot (kW)'].apply(categorize_pv)
```

```python
df_load_mean = df_load_mean.reset_index()
df_load_type = df_load_mean[['date', 'load_type']]
df_load_type
```

```python
df_pv_mean = df_pv_mean.reset_index()
df_pv_type = df_pv_mean[['date', 'solar_type']]
df_pv_type
```

# Batch

```python
load = pd.read_csv('data/load_data_15minresample_concat.csv',parse_dates=[0]).rename(columns={'Pa (kW)':'PLa',
                                                                                         'Pb (kW)':'PLb',
                                                                                         'Pc (kW)':'PLc',
                                                                                         'Ptot (kW)':'PLtot'	
                                                                                         })
pv = pd.read_csv('data/pv_data_15minresample_concat.csv',parse_dates=[0]).rename(columns={'Pa (kW)':'PVa',	
                                                                                     'Pb (kW)':'PVb',
                                                                                     'Pc (kW)':'PVc',
                                                                                     'Ptot (kW)':'PVtot'	
                                                                                     })
```

```python
pv.info()
```

```python
load.info()
```

```python
start = max(load['datetime'].min(),pv['datetime'].min()) 
end = start + pd.Timedelta(9, unit='D')
```

```python
#for 15 min sampling

num_day = 4 # unit : day
resolution = 15 # unit : min
num_interval_per_day = int(24*60/resolution)
num_total_interval = num_day * num_interval_per_day

# there are six types of batch dataset
type1 = 0
type2 = 0
type3 = 0
type4 = 0
type5 = 0
type6 = 0

start = max(load['datetime'].min(),pv['datetime'].min()) 
end = min(load['datetime'].max(),pv['datetime'].max())

dataset_detail = []
for i in range((end - start).days):
    today = start + pd.Timedelta(i,unit='D')
    next_4_days = today + pd.Timedelta(4,unit='D')
    load_batch = load[(today <=load['datetime']) & (load['datetime'] < next_4_days)][['datetime','PLa','PLb','PLc','PLtot']]
    pv_batch = pv[(today <= pv['datetime']) & (pv['datetime'] < next_4_days)][['datetime','PVa','PVb','PVc','PVtot']]
    
    # empty list for load_type and pv_type for each day
    load_type_list = []
    pv_type_list = []
    
    if load_batch.shape[0] == num_total_interval and pv_batch.shape[0] == num_total_interval:
        df_merge =  load_batch.merge(pv_batch,on = 'datetime',how = 'inner')    
        
        # create date_range for each batch    
        date_range = pd.date_range(start=today, periods=num_day, freq='D')

        # create load_type and pv_type for each day
        for day in date_range:
            type_load = df_load_type[df_load_type['date'] == day.date()]['load_type'].values[0]
            type_pv = df_pv_type[df_pv_type['date'] == day.date()]['solar_type'].values[0]
            
            load_type_list.append(type_load)
            pv_type_list.append(type_pv)
            
        data_small_batch = {'date': date_range,
                            'load_type': load_type_list,
                            'pv_type': pv_type_list}
        df_small_batch = pd.DataFrame(data_small_batch)
        
        load_count = df_small_batch['load_type'].value_counts()
        pv_count = df_small_batch['pv_type'].value_counts()
        
        load_type = load_count.idxmax()
        pv_type = pv_count.idxmax()
        
        # categorize batch dataset
        if load_type == 'high_load' and pv_type == 'high_solar':
            type1 += 1
            k = type1
            
        elif load_type == 'high_load' and pv_type == 'low_solar':
            type2 += 1
            k = type2
            
        elif load_type == 'medium_load' and pv_type == 'high_solar':
            type3 += 1
            k = type3
            
        elif load_type == 'medium_load' and pv_type == 'low_solar':
            type4 += 1
            k = type4
        
        elif load_type == 'low_load' and pv_type == 'high_solar':
            type5 += 1
            k = type5
        
        elif load_type == 'low_load' and pv_type == 'low_solar':
            type6 += 1
            k = type6  
        
        load_batch_mean = df_merge['PLtot'].mean()
        pv_batch_mean = df_merge['PVtot'].mean()
        
        df_merge.to_csv('data/batch_dataset_15min/' + load_type + '_' + pv_type + '_' + str(k) + '.csv',index=False)
        
        # prepare dataset_detail
        dataset_detail.append([load_type + '_' + pv_type + '_' + str(k), 
                               load_type, pv_type, 
                               load_batch_mean, pv_batch_mean, 
                               date_range[0], date_range[-1]])

# create dataframe of dataset_detail
dataset_detail = pd.DataFrame(dataset_detail, columns=['name','load_type','solar_type',
                                                      'load_batch_mean','pv_batch_mean',
                                                      'start_date', 'end_date'])
dataset_detail.to_csv('data/batch_dataset_15min/' + 'dataset_detail.csv',index=False)
dataset_detail
        
```
