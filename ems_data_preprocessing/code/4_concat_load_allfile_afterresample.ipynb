# After each file is resampled, concat all files together

The concat files are resampled for 5 minutes, and 15 minutes resolution. (The original is too big.)

```python
import os
import pandas as pd
```

```python
month_list = ['January2024', 'February2024']

df_5min_concat = pd.DataFrame()
df_15min_concat = pd.DataFrame()

for month in month_list:
    folder_path_5min = f"/Users/nattapongkongkaew/Desktop/input_data/preprocessed_data/data_load_5minresample/{month}/"
    folder_path_15min = f"/Users/nattapongkongkaew/Desktop/input_data/preprocessed_data/data_load_15minresample/{month}/"

    for file in os.listdir(folder_path_5min):
        if file.endswith(".csv"):
            file_path_5min = folder_path_5min + file
            df_5min = pd.read_csv(file_path_5min)
            df_5min_concat = pd.concat([df_5min_concat, df_5min])
            
    for file in os.listdir(folder_path_15min):
        if file.endswith(".csv"):
            file_path_15min = folder_path_15min + file
            df_15min = pd.read_csv(file_path_15min)
            df_15min_concat = pd.concat([df_15min_concat, df_15min])
```

```python
df_5min_concat = df_5min_concat.sort_values(by='datetime')
df_5min_concat = df_5min_concat.reset_index(drop=True) # reset index
df_5min_concat
```

```python
# Check the row count
len(df_5min_concat['datetime'].unique())
```

```python
# Save file. (Here just save the concat file to the Desktop.)
df_5min_concat.to_csv("/Users/nattapongkongkaew/Desktop/load_data_5minresample_concat.csv", index=False)
```

```python
df_15min_concat = df_15min_concat.sort_values(by='datetime')
df_15min_concat = df_15min_concat.reset_index(drop=True) # reset index
df_15min_concat
```
