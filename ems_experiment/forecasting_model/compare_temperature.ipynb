```python
import pandas as pd
import numpy as np
```

```python
def read_ncep_data(filename):
    _df = pd.read_csv(filename,sep=";", skiprows=26, names=col_names)
    _df['Time'] = _df['Time'].replace("24:00", "00:00")

    # Concatenate Date and Time columns
    _df['Datetime'] = _df['Date'] + " " + _df['Time']

    # Convert to datetime type
    _df['Datetime'] = pd.to_datetime(_df['Datetime'])

    # Identify rows where 'Time' is "00:00" and adjust the 'Date' by adding 1 day

    next_day_mask = _df['Time'] == "00:00"
    _df.loc[next_day_mask, 'Datetime'] += pd.DateOffset(days=1)
    _df['Datetime'] = _df['Datetime'] + pd.DateOffset(hours=7)

    dropped_cols = ['Date', 'Time']
    _df.drop(columns=dropped_cols, inplace=True)
    _df.set_index('Datetime', inplace=True)
    return _df
```

```python
# read NCEP data
col_names = ['Date', 'Time','Tncep', 'RH', 'Pressure', 'WS', 'WD', 'rain_fall', 'snow_fall', 'snow_depth', 'Incep']


ncep_df_2 = read_ncep_data('/content/drive/MyDrive/data for senior prj/siteCU04_2023-01-01to2023-12-31.csv') #import 2023 data
ncep_df_2['Tncep'] = ncep_df_2['Tncep']-273.5
ncep_df_2['Incep'] = ncep_df_2['Incep'] * 4
ncep_df =ncep_df_2.reset_index()
ncep_df
```

```python
vc_df = pd.read_csv('/content/drive/MyDrive/data for senior prj/visual_crossing_hourly_20230101_20231231.csv',parse_dates=['datetime'])
vc_df
```

```python
ee_df = pd.read_csv('/content/drive/MyDrive/data for senior prj/EE_Station_20230101_20231231.csv',parse_dates=['Datetime'])
ee_df
```

```python
ncep_df = ncep_df.rename(columns = {'Datetime':'datetime',})
vc_df = vc_df.rename(columns = {'temp':'T_vc','solarradiation':'I_vc'})
ee_df = ee_df.rename(columns = {'Datetime':'datetime','Ambient temperature_29 (degree C)':'T_ee','Irradiance_30 (W/m2)':'I_ee'})
```

```python
df = pd.DataFrame()
df['datetime'] = pd.date_range('2023-01-01','2023-12-31',freq='1H')
```

```python
df = df.merge(ncep_df[['datetime','Tncep']],on='datetime',how='left')
df = df.merge(vc_df[['datetime','T_vc']],on='datetime',how='left')
df = df.merge(ee_df[['datetime','T_ee']],on='datetime',how='left')
```

```python
df = df[df['Tncep']>10].dropna()
df
```

```python
def mbe(y_true,y_pred) :
    return np.mean(y_pred - y_true)
def nmae(y_true,y_pred,base='minmax') :
    if base == 'mean' :
      norm_base = y_true.mean() # mean base
    elif base == 'minmax':
      norm_base = y_true.max() - y_true.min() #min max base

    return np.mean(np.abs(y_pred - y_true))/norm_base
```

##Whole Temp Error

```python
from sklearn import metrics
mae_list = [metrics.mean_absolute_error(df['T_ee'],df['Tncep']) , metrics.mean_absolute_error(df['T_ee'],df['T_vc'])]
rmse_list = [np.sqrt(metrics.mean_squared_error(df['T_ee'],df['Tncep']) ),np.sqrt( metrics.mean_squared_error(df['T_ee'],df['T_vc']))]
mape_list = [metrics.mean_absolute_percentage_error(df['T_ee'],df['Tncep']) , metrics.mean_absolute_percentage_error(df['T_ee'],df['T_vc'])]
mbe_list = [mbe(df['T_ee'],df['Tncep']), mbe(df['T_ee'],df['T_vc'])]
nmae_list = [nmae(df['T_ee'],df['Tncep']),nmae(df['T_ee'],df['T_vc'])]
source = ['NCEP','Visual crossing']
error_value = pd.DataFrame({'source':source,'mae (kW)':mae_list,'rmse (kW)':rmse_list,'mape (%)':mape_list,'mbe (kW)':mbe_list,'nmae':nmae_list})
error_value
```

## Error by hour

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
error_df = pd.DataFrame({'hour of target':[f'{hour:02}:00' for hour in range(24)]})
hourly_mae = []
hourly_mape = []
hourly_rmse = []
hourly_mbe = []
hourly_nmae = []
for hour in range(24) :
  hour_df = df[df['datetime'].dt.hour == hour]
  hourly_mae.append([mean_absolute_error(hour_df['T_ee'],hour_df['T_vc']) ,mean_absolute_error(hour_df['T_ee'],hour_df['Tncep'])] )
  hourly_mape.append( [mean_absolute_percentage_error(hour_df['T_ee'],hour_df['T_vc'])*100, mean_absolute_percentage_error(hour_df['T_ee'],hour_df['Tncep'])*100 ])
  hourly_rmse.append([np.sqrt( mean_squared_error(hour_df['T_ee'],hour_df['T_vc']) ), np.sqrt( mean_squared_error(hour_df['T_ee'],hour_df['Tncep']) )] )
  hourly_mbe.append( [mbe(hour_df['T_ee'],hour_df['T_vc']),  mbe(hour_df['T_ee'],hour_df['Tncep'])] )
  hourly_nmae.append( [nmae(hour_df['T_ee'],hour_df['T_vc']),  nmae(hour_df['T_ee'],hour_df['Tncep'])])
error_df[['MAE_Visual Crossing','MAE_NCEP']] = hourly_mae
error_df[['MAPE_Visual Crossing','MAPE_NCEP']] = hourly_mape
error_df[['RMSE_Visual Crossing','RMSE_NCEP']] = hourly_rmse
error_df[['MBE_Visual Crossing','MBE_NCEP']] = hourly_mbe
error_df[['NMAE_Visual Crossing','NMAE_NCEP']] = hourly_mae




```
